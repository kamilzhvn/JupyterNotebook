# Loading data into pandas (task a)

```python
# Import the pandas library with the usual "pd" shortcut
import pandas as pd

import matplotlib.pyplot as plt 

df = pd.read_csv('pokemon_data.csv')
print(df.head(4))
```

## Reading data in pandas (task b)

```python
## Read headers
print(df.columns)

## Read each Column
print(df[['Name', 'Type 1', 'Attack']])

## Read Each Row
print(df.iloc[0:3])


## Read a specific location (Raw,Column)
print(df.iloc[2,1])
```

## Sorting and describing data

```python
df.describe()
# As shown in the output image, the behaviour of describe() FUNCTION  is different with series of strings.
# Statistical description of dataframe was returned with the respective passed percentiles. For the columns with strings, NaN was returned for numeric operations.
```

```python
# sorting data by name from last, taking that false count ascending 
df.sort_values('Name', ascending=False)
```

## Saving data to csv file(exporting)

```python
df.to_csv('modified_pokemon.csv')
```

## Exploring the data (task c)

```python
df['Name'].value_counts()
```

## Filtering Data 

```python
new_df = pd.read_csv('modified_pokemon.csv')
new_df = df.loc[(df['Type 1'] == 'Rock') & (df['Type 2'] == 'Fairy') & (df['HP'] > 40)]

##reset index 
new_df.reset_index( inplace=True)
print(new_df)
```

## Statistics using Groupby

```python
df = pd.read_csv('modified_pokemon.csv')
df.groupby(['Type 1']).mean().sort_values('HP', ascending = False)
```

## Working with a large amount of data

```python
for df in pd.read_csv('modified_pokemon.csv', chunksize = 5):
    print ('CHUNK DF')
    print (df)
```
